```python
import pandas as pd
```

```python
from urllib.request import urlretrieve
urlretrieve('https://hub.jovian.ml/wp-content/uploads/2020/09/countries.csv', 
            'countries.csv') # 保存到本地
```

```python
countries_df = pd.read_csv('countries.csv')
```

```python
countries_df
```

有多少个国家

```python
num_countries = countries_df.shape[0]
```

```python
num_countries
```

全世界的总人口

```python
total_population = countries_df.population.sum()
```

```python
total_population
```

全世界的平均预期寿命

```python
overal_life = (countries_df.life_expectancy*countries_df.population).sum()/total_population    
```

```python
overal_life
```

10 个 最高人口的国家

```python
most_population_df = countries_df.sort_values('population',ascending=False)
```

```python
most_population_df
```

```python
most_population_df.head(10)
```

添加一个column ,求GDP  总和

```python
countries_df['gdp'] = countries_df.population * countries_df.gdp_per_capita  
```

```python
countries_df.sort_values('gdp',ascending=False)
```

计算每个 洲  的国家数

```python
country_counts_df = countries_df.groupby('continent')['location'].count() 
country_counts_df
```

每个洲 的总人口

```python
continent_populations_df = countries_df.groupby('continent')['population'].sum() 
continent_populations_df
```
